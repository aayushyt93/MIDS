```python
# Rohan Sanjay Roman, Roll No - 63
# Batch A3

# Import required libraries
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split

# Sample dataset
documents = [
    "I love this product, it is amazing",
    "This is the best experience ever",
    "I am very happy with the service",
    "I hate this item, it is terrible",
    "Worst experience, very bad service",
    "I am disappointed and unhappy"
]

# Labels: 1 = Positive, 0 = Negative
labels = [1, 1, 1, 0, 0, 0]

# Convert text data into numeric form
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

# Split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2)

# Train Naive Bayes model
model = MultinomialNB()
model.fit(X_train, y_train)

# Test with new document
new_doc = ["The product is very good and I am happy"]
new_doc_vector = vectorizer.transform(new_doc)

prediction = model.predict(new_doc_vector)

# Output result
if prediction[0] == 1:
    print("The document is Positive")
else:
    print("The document is Negative")
```
